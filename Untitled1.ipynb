### Step 6 | Statistical Data Summary

```python
import pandas as pd
import numpy as np
from scipy import stats

def calculate_statistics(df, opacity=0.4, font_size='19px'):
    """
    Calcula estadísticas descriptivas completas y las muestra con estilo
    similar a la imagen proporcionada
    """
    # Seleccionar solo columnas numéricas
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    
    # Diccionario para almacenar estadísticas
    stats_dict = {}
    
    for col in numeric_cols:
        # Calcular todas las estadísticas
        data = df[col].dropna()  # Eliminar valores nulos para cálculos
        
        stats_dict[col] = {
            'Mean': data.mean(),
            'Median': data.median(),
            'Mode': data.mode().iloc[0] if not data.mode().empty else np.nan,
            'Range': data.max() - data.min(),
            'Stdev': data.std(),
            'Variance': data.var(),
            'IQR': data.quantile(0.75) - data.quantile(0.25),
            '25%': data.quantile(0.25),
            '50%': data.quantile(0.50),
            '75%': data.quantile(0.75),
            'Min': data.min(),
            'Max': data.max(),
            'Skewness': data.skew(),
            'Kurtosis': data.kurtosis()
        }
    
    # Crear DataFrame con las estadísticas
    stats_df = pd.DataFrame(stats_dict)
    
    # Función para formatear números
    def format_number(val):
        if pd.isna(val):
            return 'N/A'
        elif abs(val) >= 1000000:
            return f'{val:,.2f}'
        elif abs(val) >= 1000:
            return f'{val:,.2f}'
        elif abs(val) >= 1:
            return f'{val:.2f}'
        else:
            return f'{val:.2f}'
    
    # Aplicar formato a todo el DataFrame
    formatted_df = stats_df.applymap(format_number)
    
    # Definir función de estilo para filas alternas
    def highlight_rows(s):
        styles = []
        for i in range(len(s)):
            if i % 2 == 0:
                styles.append(f'background-color: rgba(173, 216, 230, {opacity}); font-weight: bold')
            else:
                styles.append(f'background-color: rgba(240, 248, 255, {opacity}); font-weight: bold')
        return styles
    
    # Aplicar estilo completo
    styled_df = formatted_df.style.apply(highlight_rows, axis=0) \
                                 .set_table_styles([
                                     # Estilo del encabezado
                                     {'selector': 'th', 'props': [
                                         ('background-color', '#4682B4'),
                                         ('color', 'white'),
                                         ('font-weight', 'bold'),
                                         ('text-align', 'center'),
                                         ('padding', '10px'),
                                         ('border', '2px solid #2F4F4F'),
                                         ('font-size', font_size)
                                     ]},
                                     # Estilo de las celdas
                                     {'selector': 'td', 'props': [
                                         ('text-align', 'center'),
                                         ('padding', '8px'),
                                         ('border', '1px solid #B0C4DE'),
                                         ('font-size', font_size),
                                         ('color', '#2F4F4F')
                                     ]},
                                     # Estilo del índice (nombres de estadísticas)
                                     {'selector': '.row_heading', 'props': [
                                         ('background-color', '#E6E6FA'),
                                         ('font-weight', 'bold'),
                                         ('text-align', 'left'),
                                         ('padding', '8px'),
                                         ('border', '1px solid #B0C4DE'),
                                         ('font-size', font_size),
                                         ('color', '#4B0082')
                                     ]},
                                     # Estilo general de la tabla
                                     {'selector': '', 'props': [
                                         ('border-collapse', 'collapse'),
                                         ('margin', '20px auto'),
                                         ('box-shadow', '0 4px 8px rgba(0,0,0,0.1)')
                                     ]}
                                 ]) \
                                 .set_properties(**{
                                     'font-family': 'Arial, sans-serif'
                                 })
    
    return styled_df

# Ejemplo de uso:
# Asumiendo que tienes un DataFrame llamado 'df' y quieres establecer transparencia a 0.4
stats_df = calculate_statistics(df, opacity=0.4, font_size='19px')
stats_df
```

```python
import pandas as pd
import numpy as np

def variable_summary(df):
    """
    Genera un resumen de variables del DataFrame con estilo similar a la imagen
    """
    # Crear lista para almacenar la información de cada variable
    summary_data = []
    
    for i, column in enumerate(df.columns):
        # Calcular estadísticas para cada columna
        dtype = str(df[column].dtype)
        count = df[column].count()  # Valores no nulos
        unique = df[column].nunique()  # Valores únicos
        missing = df[column].isnull().sum()  # Valores faltantes
        
        summary_data.append({
            'Variable': column,
            'Dtype': dtype,
            'Count': count,
            'Unique': unique,
            'Missing': missing
        })
    
    # Crear DataFrame con el resumen
    summary_df = pd.DataFrame(summary_data)
    
    # Función para aplicar estilo alternado
    def highlight_rows(row):
        styles = []
        if row.name % 2 == 0:
            # Filas pares - azul más claro
            styles = ['background-color: #E6E6FA; color: #4B0082'] * len(row)
        else:
            # Filas impares - azul más oscuro
            styles = ['background-color: #D8BFD8; color: #4B0082'] * len(row)
        return styles
    
    # Aplicar estilo
    styled_df = summary_df.style.apply(highlight_rows, axis=1) \
                               .set_table_styles([
                                   # Estilo del encabezado
                                   {'selector': 'th', 'props': [
                                       ('background-color', '#9370DB'),
                                       ('color', 'white'),
                                       ('font-weight', 'bold'),
                                       ('text-align', 'center'),
                                       ('padding', '12px'),
                                       ('border', '1px solid #8A2BE2'),
                                       ('font-size', '14px')
                                   ]},
                                   # Estilo de las celdas
                                   {'selector': 'td', 'props': [
                                       ('text-align', 'center'),
                                       ('padding', '10px'),
                                       ('border', '1px solid #9370DB'),
                                       ('font-size', '13px'),
                                       ('font-weight', 'bold')
                                   ]},
                                   # Estilo del índice
                                   {'selector': '.row_heading', 'props': [
                                       ('background-color', '#9370DB'),
                                       ('color', 'white'),
                                       ('font-weight', 'bold'),
                                       ('text-align', 'center'),
                                       ('padding', '10px'),
                                       ('border', '1px solid #8A2BE2'),
                                       ('font-size', '13px')
                                   ]},
                                   # Estilo general
                                   {'selector': '', 'props': [
                                       ('border-collapse', 'collapse'),
                                       ('margin', '20px auto'),
                                       ('box-shadow', '0 4px 8px rgba(0,0,0,0.2)')
                                   ]}
                               ]) \
                               .set_properties(**{
                                   'font-family': 'Arial, sans-serif'
                               })
    
    return styled_df

# Para usar la función:
# DataFrame llamado 'df'
summary_df = variable_summary(df)
summary_df
```
